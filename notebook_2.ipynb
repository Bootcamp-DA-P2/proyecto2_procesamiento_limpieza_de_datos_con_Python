### Comprobación de nulos

```python
data.isnull()
```

```python
data.isnull().mean() * 100
```

```python
data = data.drop(columns=['tags']) #Eliminamos tags debido a que tiene un 25% de nulos
data.info()
```

- Sustitución de nulos por "No especificado" en caso de string y por la media numerica de la columna en caso de número

```python

#Seleccionar los nombres de las columnas que son texto
columns_str = data.select_dtypes(str).columns
print(columns_str)

for column in columns_str:
    data[column] = data[column].fillna("No especificado")

# Seleccionamos las que son int o float
columns_num = data.select_dtypes('number').columns

for column in columns_num:
    # Calculamos la mediana de la columna
    mediana = data[column].median()
    # Rellenamos los nulos de esa columna con su mediana
    data[column] = data[column].fillna(mediana)

# Verificamos que ya no quedan nulos
print(data.isnull().sum())


```

### Normalización de textos

```python
for column in data.columns:
    if data[column].dtype == 'str':
        data[column] = data[column].str.strip().str.capitalize() #Eliminamos espacios en blanco y ponemos la primera letra en mayusculas

```

### Normalización de monedas

- Eliminación de country code debido a redundancia con la columna de country y conversión de currency a dolares para poder eliminar la misma columna de country

```python
data = data.drop(columns=['country_code']) #Eliminación de columna contry_code
```

```python
import requests

# Obtenemos tasas de cambio
response = requests.get('https://api.exchangerate-api.com/v4/latest/USD')
rates_usd_base = response.json()['rates']

#Creamos un diccionario vacio al cual decimos que recorra el de rates_usd_base para invertir los valores
rates_to_usd = {}
for currency, rate in rates_usd_base.items():
    rates_to_usd[currency] = 1 / rate

data['currency'] = data['currency'].str.upper() #normalizamos a mayusculas para que coincidan con los datos de conversion de la API

def get_rate(currency_code):
    return rates_to_usd.get(currency_code, None) # En caso de no encontrar la moneda devolveremos None


#Creamos dos nuevas columnas para meter estos nuevos datos
data['funded_amount_usd'] = data['funded_amount'] * data['currency'].apply(get_rate)
data['loan_amount_usd'] = data['loan_amount'] * data['currency'].apply(get_rate)
```

```python
#Eliminar  funded_amount y loan_amount puesto que ya no son necesarias y podemos calcularas en base a funded_amount_usd y la tasa de conversión
data = data.drop(columns=['funded_amount','loan_amount'])
```
