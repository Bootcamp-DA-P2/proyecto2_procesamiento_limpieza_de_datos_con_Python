- Sustitución de nulos por "No especificado" en caso de string y por la media numerica de la columna en caso de número

## Analisis exploratorio

1º LINEA DE ANALISIS: EVOLUCION MENSUAL Y ESTACIONALIDAD - EL FACTOR TIEMPO (date)

```python
#Aseguramos formato fecha
data['date'] = pd.to_datetime(data['date'])
#Extraemos año y mes para analisis de estacionalidad
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
```

```python
# Agrupamos por mes real
grupos_mensuales = data['date'].dt.to_period('M')
evolucion_mensual = data.groupby(grupos_mensuales).agg(
    conteo_prestamos=('funded_amount', 'count'),
    monto_total=('funded_amount', 'sum'),
    monto_promedio=('funded_amount', 'mean')
)
```

```python
# Calculamos el crecimiento porcentual mes a mes
evolucion_mensual['crecimiento_vol_pct'] = evolucion_mensual['conteo_prestamos'].pct_change() * 100
# Mostramos resultado
print(evolucion_mensual.head(12))
```

- CONCLUSION: 
El volumen de préstamos muestra una tendencia de crecimiento anual con picos máximos en el último trimestre (octubre-noviembre). Sin embargo, existe una estacionalidad negativa marcada en diciembre, donde el volumen cae drásticamente. Además, se observa que en los meses de mayor actividad, el monto promedio por préstamo tiende a disminuir, sugiriendo una mayor diversificación hacia microcréditos de menor escala.

```python
#Aquí agrupamos todos los 'meses'..etc
estacionalidad = data.groupby('month').agg(
    promedio_mensual_prestamos=('funded_amount', 'count'),
    promedio_monto_solicitado=('funded_amount', 'mean')
).sort_values('month')
print("--- EVOLUCIÓN TEMPORAL (Primeras 10 filas) ---")
print(evolucion_mensual.head(10))

print("\n--- PATRÓN POR MES (ESTACIONALIDAD) ---")
print(estacionalidad)
```

- CONCLUSION: 
El mes con  mayor volumen de prestamos es el mes de marzo con 67.583 préstamos. Podemos observar que hay un aumento de enero hasta marzo. Puede que venga motivado por el inicio de ciclos productivos o escolares en muchas regiones del mundo. A partir del mes de julio, el volumen cae significativamente y se mantiene en un nivel bajo y estable hasta final de año. Podemos observar que enero y febrero tienen los promedios de monto solicitados más altos a pesar de no ser los meses con más creéitos concedidos.

4º LINEA DE ANALISIS: ¿CUAL ES LA RELACION ENTRE LOS MONTOS? ¿HAY PRESTAMOS QUE SE QUEDAN SIN FINANCIAR?

```python
data.loc[:,['funded_amount','loan_amount']]
```

```python
no_completados = (data['funded_amount'] < data['loan_amount']).sum()

cero_financiacion = (data['funded_amount'] == 0).sum() #Contamos cuantos tienen 0 financiación

print(f"Préstamos con financiación incompleta: {no_completados}")
print(f"Préstamos con CERO financiación: {cero_financiacion}")
```

```python
pct_incompletos = (data['funded_amount'] < data['loan_amount']).mean() * 100

print(f"Porcentaje de prestamos sin completar: {pct_incompletos:.2f}%")
```

- CONCLUSION: Debido a que el porcentaje sin 7.2% podemos deducir que existe una gran demanda que la plataforma no logra cubrir.
