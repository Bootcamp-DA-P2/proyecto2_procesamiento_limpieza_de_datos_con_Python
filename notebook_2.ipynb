## Limpieza de datos

- Pasar de tipo str a datetime las fechas

```python
data_type = data.loc[:,['posted_time', 'disbursed_time', 'funded_time', 'date']]
data_type.info()
```

```python
colum_date = ['posted_time', 'disbursed_time', 'funded_time', 'date']
for column in colum_date:
    data[column] = pd.to_datetime(data[column]) #Por cada columna cambia a tipo fecha

data_type = data.loc[:,['posted_time', 'disbursed_time', 'funded_time', 'date']]
data_type.info()
```

- Comprobación de nulos

```python
data.isnull()
```

```python
data.isnull().mean() * 100
```

```python
data = data.drop(columns=['tags']) #Eliminamos tags debido a que tiene un 25% de nulos
data.info()
```

- Sustitución de nulos por "No especificado" en caso de string y por la media numerica de la columna en caso de número

```python

#Seleccionar los nombres de las columnas que son texto
columns_str = data.select_dtypes(str).columns
print(columns_str)

for column in columns_str:
    data[column] = data[column].fillna("No especificado")

# Seleccionamos las que son int o float
columns_num = data.select_dtypes('number').columns

for column in columns_num:
    # Calculamos la mediana de la columna
    mediana = data[column].median()
    # Rellenamos los nulos de esa columna con su mediana
    data[column] = data[column].fillna(mediana)

# Verificamos que ya no quedan nulos
print(data.isnull().sum())


```

```python
data = data.dropna() #borra las filas que tengan valores faltantes
data.info()
```
